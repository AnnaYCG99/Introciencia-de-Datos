```python
import pandas as pd
```

```python
data={'p':[True, True, False, False], 'q':[True, False, True, False]}
df=pd.DataFrame(data)
df2=pd.DataFrame(data)
df
```

```python
df['p^q']=[p and q for p, q in zip(df['p'], df['q'])]
df
```

```python
df['(p^q)vq']=[(p and q) or q for p, q in zip(df['p'], df['q'])]
df
```

```python
df2['(~pvq)']=[(not p or q) for p, q in zip(df['p'], df['q'])]
df2
```

```python
df2['(pv~q)']=[(p or not q) for p, q in zip(df['p'], df['q'])]
df2
```

```python
df2['(~pvq)^(pv~q)']=[(not p or q) and (p or not q) for p, q in zip(df['p'], df['q'])]
df2
```

```python
df2['~((~pvq)^(pv~q))']=[not((not p or q) and (p or not q)) for p, q in zip(df['p'], df['q'])]
df2
```

```python
df3=df2.drop(columns=['(~pvq)^(pv~q)'])
df3
```
